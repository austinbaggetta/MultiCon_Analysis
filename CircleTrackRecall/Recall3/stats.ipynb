```python
import numpy as np 
import pandas as pd 
import pingouin as pg
```

```python
## Settings
parent_dir = 'CircleTrack_Recall'
experiment_dir = 'Recall3'
ct_df = pd.read_csv('./ct_df.csv')
mouse_metrics = pd.read_csv('./mouse_metrics.csv')
scatterplot = pd.read_csv('./scatterplots.csv')
```

```python
## Lick accuracy two-way ANOVA
ct_df.mixed_anova(dv='lick_accuracy', within='day', between='group', subject='mouse')
```

```python
## Lick accuracy multiple comparisons testing due to a main effect of group
ct_df.pairwise_tests(dv='lick_accuracy', within='day', between='group', subject='mouse', padjust='fdr_bh')
```

```python
## Hit rate 2-way ANOVA
mouse_metrics.mixed_anova(dv='hits', within='day', between='group', subject='mouse')
```

```python
## Correct rejection rate 2-way ANOVA
mouse_metrics.mixed_anova(dv='CR', within='day', between='group', subject='mouse')
```

```python
## Correct rejection rate multiple comparisons
mouse_metrics.pairwise_tests(dv='CR', within='day', between='group', subject='mouse', padjust='fdr_bh')
```
